## Undirected Graph of the Sentence including a Linear Chain of tokens

```python
from IPython.display import Image
Image('/Users/simontse/Desktop/Screenshot 2022-02-04 at 9.50.13 AM.png')

# "All-paths graph kernel for protein-protein interaction extraction with evaluation of cross-corpus learning"
# from Natural Language Processing in Biomedicine (BioNLP) ACL Workshop 2008 Columbus, OH, USA. 19 June 2008, Published: 19 November 2008
# BMC Bioinformatics 2008, 9(Suppl 11):S2 doi:10.1186/1471-2105-9-S11-S2
```

```python
# Helper function to create an Undirected Graph
# The graph would be composed to two sections
# First section is a linear chain of the tokens that follows the order of the original sentence (i.e. Lower part of the image)
# Second section is a dependency path that connect tokens based on their 'head' (i.e. Upper part of the image)
# Note the edge weight is assigned as 1 instead of the <1 value specificed in the literature.

def createTokenGraphwithContractionV2(doc):
    """ 
    input: a doc generated from stanza NLP pipeline
    output: an undirected graph with contraction made to the sentence end symbol "#"
    """
    H = nx.Graph()
    
    # Add linear chain of tokens
    for sent in doc.sentences:
        ids = len(sent.words)
        for i in range(ids-1):
            head = sent.words[i].text
            head_pos = sent.words[i].upos
            head_head = sent.words[i].head
            tail = sent.words[i+1].text
            tail_pos = sent.words[i+1].upos
            tail_head = sent.words[i+1].head
            #print (head, head_pos, tail, tail_pos)
            if head_pos == 'PUNCT' or tail_pos == 'PUNCT':
                pass
            else:
                H.add_node(head, pos=head_pos, head=head_head)
                H.add_node(tail, pos=tail_pos, head=tail_head)
                H.add_edge(head, tail, weight=1)
    
    # Add chain that based on the dependency parsing result
    for sent in doc.sentences:
        for word in sent.words:
            head = sent.words[word.head-1].text
            head_pos = sent.words[word.head-1].upos 
            tail = word.text
            tail_pos = word.upos
            #print (head, '->', tail)
            if head_pos == 'PUNCT' or tail_pos=='PUNCT': # or word.text=='#' or sent.words[word.head-1].text=='#'
                pass
            else:            
                H.add_node(head, pos=head_pos)
                H.add_node(tail, pos=tail_pos)
                H.add_edge(head, tail, weight=1)
    
    # Contractipn of sentence separator "#            
    if '#' in H.nodes:
        for n in H.neighbors('#'):
            if H.nodes[n]['pos']=='VERB':
                G = nx.contracted_nodes(H, n, '#')
                return G
            else:
                return H
    else:
        return H
```

```python
h = createTokenGraphwithContractionV2(doc)
bmt.plot_network(h)
print(h.nodes)
```

```python
directed_tokens = bmt.extractTokensOfInterestfmGraph(h, entities)
directed_tokens = list(set(directed_tokens))
directed_tokens
```

```python
from itertools import combinations, permutations
print (df.loc[134,'sentence']+'\n')
length = 0
count = 0
tuples = combinations(directed_tokens, 2)
for t in tuples:
    paths = nx.all_simple_paths(h, t[0], t[1])
    for index, p in enumerate(paths):
        print (index+1,":",p)
        length += len(p)
        count += 1
print ('\n'+'Average Path Length: %.2f' %(float(length)/float(count)))
```

From above output, it's apparent adding a linear chain to the graph representation of the sentence will generate more viable paths connecting the two entities and it's likely to increase the statistical power to discern if the entities of interest are having any meaningful relationship based on number of paths and nature of in-between tokens.

```python
from itertools import combinations
print (df.loc[134,'sentence']+'\n')
length = 0
count = 0
tuples = combinations(directed_tokens, 2)
for t in tuples:
    paths = nx.all_shortest_paths(h, t[0], t[1])
    for index, p in enumerate(paths):
        print (index+1,":",p)
        length += len(p)
        count += 1
print ('\n'+'Average Path Length: %.2f' %(float(length)/float(count)))
```

From above using shortest paths, it will further condense the 'connection' between the two entities of interest.  Combining with the result from all simple paths, it's obvious that other phrases are also important such as <b>'C-terminal region'</b> to explain the relationship.
